## Preprocess Data

```python
df['확진일'] = pd.to_datetime(df['확진일'], format="%Y/%m/%d")
df['등록일'] = pd.to_datetime(df['등록일'], format="%Y/%m/%d %H:%M:%S")
df['수정일'] = pd.to_datetime(df['수정일'], format="%Y/%m/%d %H:%M:%S")
df.dtypes
```

## Create Sub-tables For Patients, Non-patients, Deaths

```python
non_patients = df.loc[df['상태']=='퇴원']
non_patients.head()
```

```python
patients = df.loc[df['상태'].isnull()]
patients.head()
```

```python
deaths = df.loc[df['상태']=='사망']
deaths.head()
```

## Convert '상태' Column Data From Object Into Int For Easy Data Handling

```python
test = df[['확진일', '상태']]
status = []

for x in test['상태']:
    if x == '퇴원':
        status.append(1)
    elif x == '사망':
        status.append(2)
    else:
        status.append(0)
test['상태']=status
test.dtypes
```

```python
a = np.sum(test['상태'] == 2)
b = np.sum(test['상태'] == 0)
c = np.sum(test['상태'] == 1)
print(a,b,c)
```

## Create Table With Column '상태' With Count Values

```python
groupby_status = test.groupby('상태')
groupby_status.size()
test2 = pd.DataFrame({'count': groupby_status.size()})
test2
```

```python
groupby_loc = df.groupby('지역')
groupby_loc.size()
loc = pd.DataFrame({'loc_count': groupby_loc.size()})
loc
```

## Plot Bar Graph

```python
# 0: 확진, 1: 퇴원, 2: 사망
# 결과: 퇴원한 사람 수가 현저히 많음
test2.plot.bar()
```
